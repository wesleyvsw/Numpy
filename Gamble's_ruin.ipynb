```
#Para 6 casos diferentes
valor = [[20],[20],[20],[20],[20],[20]]
tentativas = [[0],[0],[0],[0],[0],[0]]
serie_valor =[[20],[20],[20],[20],[20],[20]]
for x in range(6):
  while valor[x][0] != 0 and valor[x][0] != 60:
    rand = bernoulli.rvs(0.5,size=1)
    if rand[0] == 0:
      valor[x][0] = valor[x][0]-1
    else:
      valor[x][0] = valor[x][0]+1
    print(valor)
    tentativas[x][0] = tentativas[x][0] + 1
    serie_valor[x].append(valor[x][0])
print(tentativas)
```
